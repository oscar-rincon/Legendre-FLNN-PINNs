```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import time
device = torch.device('cpu')  # Usamos CPU para este ejemplo
```

```python
# Function to compute derivatives
def derivative(dy: torch.Tensor, x: torch.Tensor, order: int = 1) -> torch.Tensor:
    for i in range(order):
        dy = torch.autograd.grad(
            dy, x, grad_outputs=torch.ones_like(dy), create_graph=True, retain_graph=True
        )[0]
    return dy
```

```python
# Define a neural network class with three fully connected layers
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.layer1 = nn.Linear(1, 10, dtype=torch.float64)  # Capa con doble precisión
        self.layer2 = nn.Linear(10, 10, dtype=torch.float64)  # Capa con doble precisión
        self.output_layer = nn.Linear(10, 1, dtype=torch.float64)  # Capa con doble precisión

    def forward(self, x):
        x = torch.tanh(self.layer1(x))
        x = torch.tanh(self.layer2(x))
        x = self.output_layer(x)
        return x
    
```

```python
lb = 1e-16
ub = 1
num_samples = 10
x = torch.linspace(lb, ub, num_samples, dtype=torch.float64, requires_grad=True).unsqueeze(1).to(device)
# Create an instance of the neural network and move it to the GPU
model = NeuralNetwork().to(device)
num_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f'El modelo tiene {num_params} parámetros.')

# Definir la función de pérdida y el optimizador
criterion = nn.MSELoss()
optimizer = torch.optim.LBFGS(model.parameters(),
                                lr=1,
                                max_iter=10_000,
                                max_eval=10_000,
                                tolerance_grad=1e-6,
                                history_size=50,
                                tolerance_change=1.0 * np.finfo(float).eps,
                                line_search_fn=None)
 

# Definir la función de cierre
iter_count=0
def closure():
    global iter_count
    iter_count+=1
    optimizer.zero_grad()
    y_pred = model(x) 
    y_t = x*(x-1)*y_pred
    y_xx = derivative(y_t, x, order=2)
    residuo =  y_xx +  y_t + x
    loss = criterion(residuo,torch.zeros_like(residuo))
    #if iter_count % 100 == 0: 
    #    print(f'iter {iter_count}, loss: {loss.detach().numpy()}')
    loss.backward(retain_graph=True)
    return loss

# Función para entrenar el modelo
def train_model():
    start_time = time.time()
    optimizer.step(closure)
    end_time = time.time()
    y_pred = model(x) 
    y_t = x*(1-x)*y_pred
    y_xx = derivative(y_t, x, order=2)
    residuo =  y_xx +  y_t + x
    loss = criterion(residuo,torch.zeros_like(residuo)) 
    print(f'Final loss: {loss.detach().numpy()}')
    print(f'Tiempo de entrenamiento: {end_time - start_time} segundos')

# Entrenar el modelo
train_model()
```

```python
# Define input x with double precision
x = torch.tensor([0.5191], dtype=torch.float64)

# Ensure prediction y_pred also uses double precision
with torch.no_grad():
    #input_tensor = torch.tensor([1.1320], dtype=torch.float64).to(device)
    y_pred = model(x)

# Perform the computation with double precision
y_pred = x*(1-x)*y_pred  

# Print the result
print(np.round(y_pred.item(), 4))
```

```python
# Define the analytical function
def analytical(x):
    return np.sin(x)/np.sin(1)-x

x_values = np.linspace(lb, ub, 10)

# Calculate y_pred using your model
x_tensor = torch.linspace(lb, ub, num_samples, dtype=torch.float64, requires_grad=True).unsqueeze(1).to(device)
with torch.no_grad():
    y_pred = model(x_tensor)
y_pred = x_tensor*(x_tensor-1) * y_pred

# Convert tensors to numpy arrays for visualization
x_values = x_values
y_pred = y_pred.cpu().detach().numpy()

# Calculate analytical values for comparison
y_analytical = analytical(x_values)

# Plot the results
plt.plot(x_values, y_pred, label='Model Prediction')
plt.plot(x_values, y_analytical, label='Analytical Solution', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparativa entre Solución Analítica y Predicción del Modelo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the results
plt.plot(x_values, y_analytical-y_pred[:,0])
plt.xlabel('x')
plt.ylabel('y')
plt.ylim([-1e-5,1e-5])
plt.title('Error')
plt.legend()
plt.grid(True)
plt.show()
```
